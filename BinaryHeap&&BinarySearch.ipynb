```python
def binarySearch(alist,item):
    first=0
    last=len(alist)-1
    
    while first<=last:
        mid=(first+last)//2
        if alist[mid]==item:
            return mid+1
        elif alist[mid]>item:
            last=mid-1
        else:
            first=mid+1
    return -1

test=[0,1,2,8,13,17,19,32,42]
print(binarySearch(test,17))
        
```
